```python
import sys
from pathlib import Path

# Make project root importable so `from src...` works
project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

import json
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)
```

```python
BOOTSTRAP_PATH = project_root / "data" / "raw" / "bootstrap_static.json"
FIXTURES_PATH  = project_root / "data" / "raw" / "fixtures.json"

assert BOOTSTRAP_PATH.exists(), f"Missing: {BOOTSTRAP_PATH}. Run: python -m src.fpl.fetch"
assert FIXTURES_PATH.exists(), f"Missing: {FIXTURES_PATH}. Run: python -m src.fpl.fetch"

with open(BOOTSTRAP_PATH, "r") as f:
    bootstrap = json.load(f)

with open(FIXTURES_PATH, "r") as f:
    fixtures_raw = json.load(f)

players = pd.DataFrame(bootstrap["elements"])
teams   = pd.DataFrame(bootstrap["teams"])
events  = pd.DataFrame(bootstrap["events"])

fixtures = pd.DataFrame(fixtures_raw)

(players.shape, teams.shape, events.shape, fixtures.shape)
```

```python
pos_map = {1: "GK", 2: "DEF", 3: "MID", 4: "FWD"}
team_id_to_name = dict(zip(teams["id"], teams["name"]))

players["position"] = players["element_type"].map(pos_map)
players["team_name"] = players["team"].map(team_id_to_name)

# Cost is stored as int tenths (e.g., 75 -> £7.5)
players["price"] = players["now_cost"] / 10.0

# Some columns are strings like "12.3" — convert where needed
players["selected_by_percent"] = pd.to_numeric(players["selected_by_percent"], errors="coerce")
players["form"] = pd.to_numeric(players["form"], errors="coerce")
players["points_per_game"] = pd.to_numeric(players["points_per_game"], errors="coerce")

players[["web_name","position","team_name","price","minutes","total_points","points_per_game","form","selected_by_percent","status"]].head()
```

```python
players_active = (
    players
    .query("total_points > 0 and status == 'a'")
    .sort_values("form_rank", ascending=True)
    .reset_index(drop=True)
)

players_active.shape
```
